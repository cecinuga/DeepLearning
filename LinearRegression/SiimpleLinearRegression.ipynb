```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
```

```python
X, y = datasets.make_regression(n_samples=200, n_features=1, n_targets=1, random_state=45, noise=25)
y = np.array([[r] for r in y])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

fig, ax = plt.subplots(figsize=(4, 4))
ax.scatter(X[:,0], y[:, 0])
plt.show()
```

```python
def loss_function(m, b, points):
    return np.sum([y - (m * x + b) ** 2 for x, y in points])/float(len(points))

def gradient_descent(m_now, b_now, points, L):
    m_gradient = np.sum([[-(2/len(points)) * x * (y - m_now * x + b_now)] for x, y in points])
    b_gradient = np.sum([[-(2/len(points)) * (y - (m_now * x + b_now))] for x, y in points])
    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    return m, b
```

```python
m = 0
b = 0
L = 0.001
epochs = 1000
points = np.concatenate([X, y], dtype=np.float64, axis=1)
x = np.linspace(-4, 4, 1000)
for i in range(epochs):
    m, b = gradient_descent(m, b, points, L)

print(m, b)
plt.scatter(X, y)
plt.plot(x, [m * i + b for i in x])
```
